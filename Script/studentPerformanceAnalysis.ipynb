# Calculating Semester average

```python
df['sem_Avg']=(df['SGPA_SEMESTER1']+df['SGPA_SEMESTER2']+df['SGPA_SEMESTER3']+df['SGPA_SEMESTER4'])/4
df.head()
```

```python
df['FIELD_OF_INTEREST'].unique()
```

```python
for i in range(len(df['FIELD_OF_INTEREST'])):
    if(df['FIELD_OF_INTEREST'][i]=='AERUNAUTICS'):
        df['FIELD_OF_INTEREST'] = df['FIELD_OF_INTEREST'].str.replace("AERUNAUTICS","FRONT END")
for i in range(len(df['FIELD_OF_INTEREST'])):
    if(df['FIELD_OF_INTEREST'][i]=='MUSIC'):
        df['FIELD_OF_INTEREST'] = df['FIELD_OF_INTEREST'].str.replace("MUSIC","FRONT END")
```

```python
my_dict={
    'ENGINEERING':1, 
    'CODING':2, 
    'TESTING':3,
    'DEVELOPMENT':4,
    'RESEARCH':5,
    'AUTOMATION':6, 
    'DESIGNING':7, 
    'TEACHING':8,
    'BACK END':9,
    'FRONT END':10, 
    'DANCE':11, 
    'PAINTING':12, 
    'PHYSCIOLOGY':13, 
    'SPORTS':14, 
    'FASHION DESIGNING':15
}
df['SCORE_INTEREST']=1
my_dict['ENGINEERING']
for i in range(0,len(df['FIELD_OF_INTEREST'])):
    score =my_dict[df['FIELD_OF_INTEREST'][i]]
    df['SCORE_INTEREST'][i]=score
```

```python
df['IMPROVEMENT_SCORE']=0
for i in range(len(df['10 (%)'])):
    if((df['10 (%)'][i]<df['12 (%)'][i]) and (df['SGPA_SEMESTER1'][i]<df['SGPA_SEMESTER2'][i]) and (df['SGPA_SEMESTER2'][i]<df['SGPA_SEMESTER3'][i]) and (df['SGPA_SEMESTER3'][i]<df['SGPA_SEMESTER4'][i])):
        df['IMPROVEMENT_SCORE'][i]=1
    elif((df['10 (%)'][i]>df['12 (%)'][i]) and (df['SGPA_SEMESTER1'][i]<df['SGPA_SEMESTER2'][i]) and (df['SGPA_SEMESTER2'][i]<df['SGPA_SEMESTER3'][i]) and (df['SGPA_SEMESTER3'][i]<df['SGPA_SEMESTER4'][i])):
        df['IMPROVEMENT_SCORE'][i]=1
    elif((df['10 (%)'][i]<df['12 (%)'][i]) and not((df['SGPA_SEMESTER1'][i]<df['SGPA_SEMESTER2'][i]) and (df['SGPA_SEMESTER2'][i]<df['SGPA_SEMESTER3'][i]) and (df['SGPA_SEMESTER3'][i]<df['SGPA_SEMESTER4'][i]))):
        df['IMPROVEMENT_SCORE'][i]=0
    elif((df['10 (%)'][i]<df['12 (%)'][i]) and ((df['SGPA_SEMESTER1'][i]<df['SGPA_SEMESTER2'][i]) or ((df['SGPA_SEMESTER2'][i]<df['SGPA_SEMESTER3'][i]) or ((df['SGPA_SEMESTER3'][i]<df['SGPA_SEMESTER4'][i]))))):
        df['IMPROVEMENT_SCORE'][i]=1
    elif((df['10 (%)'][i]<df['12 (%)'][i]) or (df['SGPA_SEMESTER1'][i]<df['SGPA_SEMESTER2'][i]) and (df['SGPA_SEMESTER2'][i]<df['SGPA_SEMESTER3'][i]) and (df['SGPA_SEMESTER3'][i]<df['SGPA_SEMESTER4'][i])):
        df['IMPROVEMENT_SCORE'][i]=1
```

```python
import numpy as np
```

```python
dummy2=dummy2.replace([np.inf, -np.inf], np.nan).dropna(axis=1)
```

```python
targets=targets.replace([np.inf, -np.inf], np.nan).dropna(axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dummy2, targets, test_size=0.20, random_state=42)
```

```python
print("The shape of the training data and labels are, {} and {}".format(X_train.shape, y_train.shape))
print("\n")
print("The shape of the training data and labels are, {} and {}".format(X_test.shape, y_test.shape))
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
RFReg =RandomForestRegressor(n_estimators = 10, random_state = 42)
RFReg.fit(X_train, y_train)
```

```python
from sklearn import metrics
```

```python
y_predict_rfr = RFReg.predict((X_test))
r_square = metrics.r2_score(y_test,y_predict_rfr)
print('R-square Error Associated with Random Forest Regression is:', r_square)
```
